# How to create custom calendar

```python
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday

class myBirthDayCalendar(AbstractHolidayCalendar):
    rules = [
        Holiday('Kushal bday', month = 6, day = 7)
    ]
myc = CustomBusinessDay(calendar = myBirthDayCalendar())
myc
```

```python
pd.date_range(start = '6/1/2019', end='6/30/2019', freq=myc)
```

### In egypt weekend is Friday & Saturday so this can be cleared using weekmask

```python
b = CustomBusinessDay(weekmask = 'Sun Mon Tue Wed Thu')
pd.date_range(start = '3/1/2019', end='3/1/2019', freq='B')
```
