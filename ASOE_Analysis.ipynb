### Extracting month and year from MonthDayYearProjectStartDate column

```python
# Date Format
df["StartDate"]=pd.to_datetime(df["MonthDayYearofProjectStartDate"], errors='coerce')
df["Month"]=df["MonthDayYearofProjectStartDate"].dt.month
df["Year"]=df["MonthDayYearofProjectStartDate"].dt.year
df["Quarter"]=df["MonthDayYearofProjectStartDate"].dt.quarter

df['YearMonth']  = df['Year'].astype(str) + "-" +df['Month'].astype(str)
df['YearQuarterCal']  = df['Year'].astype(str) + "-" +df['Quarter'].astype(str)

df["StartDate"].head(1), df["Month"].head(1), df["Year"].head(1), df["Quarter"].head(1), df["YearMonth"].head(1), df["YearQuarterCal"].head(1)
```
